```python
import pandas as pd
import csv

from bs4 import BeautifulSoup
from urllib.request import urlopen
path = "11"
f = open(fr"C:/Memoticon/tmp_html/{path}.html", "r", encoding="utf8")
soup = BeautifulSoup(f, "html.parser")
```

```python
#회사 목록 세기
count = len(soup.find_all('tr', class_="devloopArea"))
count
```

```python
#회사 한줄 가져오기
company = soup.find_all('tr', class_="devloopArea")

#리스트 설정
company_name_list = []
job_position_list = []
job_career_list = []
job_education_list = []
job_location_list = []
job_date_list = []

#for i in range(0,count):
for i in range(0,24):
    try : 
        print()
        print("----------------------------------------",i)
        soup2 = company[i]

        #회사 이름 추출
        company_name = soup2.find_all('td', class_="tplCo")
        company_name = soup2.find_all('a', class_="link normalLog")
        company_name = company_name[0].get_text().strip()
        company_name_list.append(company_name)
        print(company_name)

        #채용 분야 추출
        job_information = soup2.find_all('td', class_="tplTit")
        job_position = soup2.find_all('a', class_="link normalLog")
        job_position = job_position[1].get_text().strip()
        job_position_list.append(job_position)
        print(job_position)

        #경력
        job_information = soup2.find_all('td', class_="tplTit")
        job_additional = soup2.find_all('span', class_="cell")
        job_career = job_additional[0].get_text().strip()
        job_career_list.append(job_career)
        print(job_career )

        #학력
        job_information = soup2.find_all('td', class_="tplTit")
        job_additional = soup2.find_all('span', class_="cell")
        job_education = job_additional[1].get_text().strip()
        job_education_list.append(job_education)
        print(job_education)

        #지역
        job_information = soup2.find_all('td', class_="tplTit")
        job_location = job_additional[2].get_text().strip()
        job_location_list.append(job_location)
        print(job_location )

        #마감일자
        job_date = soup2.find_all('td', class_="odd")
        job_date = soup2.find_all('span', class_="tahoma")
        job_date = job_date[1].get_text().strip()
        job_date_list.append(job_date)
        print(job_date)
        
    except:    # 예외가 발생했을 때 실행됨
        print('Error')
        job_date = "오늘 또는 내일 마감"
        job_date_list.append(job_date)
        print(job_date)
```

```python
# 개별 리스트를 데이터프레임으로 병합

job_df = pd.DataFrame({'회사명':company_name_list, '직무':job_position_list, '경력':job_career_list, '학력':job_education_list, '지역':job_location_list, '마감일':job_date_list})
```
