```python
for idx, value in enumerate(pinterest.description):
    if "|" in str(value):
        pinterest.description[idx] = value.split("|")[-1]
    elif value == "":
        pinterest.description[idx] = "NaN"
```

```python
tag = []
for i in pinterest["title"]:
    try :
        tag.append(okt.phrases(i))
    except:
        tag.append(None)
pinterest['tag'] = tag
```

```python
tag = []
for i in pinterest["title"]:
    try :
        tag.append(okt.morphs(i))
    except:
        tag.append(None)
pinterest['tag2'] = tag
```

```python
for i, value in enumerate(pinterest["description"]):
    if pinterest['tag'][i] == None:
        try:
            pinterest['tag'][i] = okt.phrases(value)
        except:
            pass
    else:
        pass
```
